```python
(_filter := df.loc[:, 'revenue'] > df.revenue.quantile(0.5))
```

```python
df.loc[_filter]
```

```python
filter_1 = df.revenue > df.revenue.quantile(0.5)
filter_2 = df.operatingCashFlow > df.operatingCashFlow.quantile(0.1)
filter_3 = df.debtToEquity < df.debtToEquity.quantile(0.1)
list(map(lambda x: x.shape, [df[filter_1], df[filter_2], df[filter_3]]))  # print chosen # of stocks for each filter
```

```python
df.loc[filter_1 & filter_2 & filter_3]
```
